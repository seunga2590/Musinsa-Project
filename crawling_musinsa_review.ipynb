```python
# 필요한 패키지
import requests #urllib, request...
import urllib.request
import json

# 데이터 처리
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup

# 기타 필요 : 정규식, 시간
import re #정규식
import time
```

```python
reviews = []

page = '0'
goodsNo = '1434514'


while True:
    try:
        page = str(int(page)+1)
        url = 'https://goods.musinsa.com/api/goods/v2/review/goods/list?similarNo=0&sort=new&selectedSimilarNo=0&page=%s&goodsNo=%s&_=1668584129114' % (page, goodsNo)
        req = urllib.request.Request(url,
                                    headers = {"referer": "https://www.musinsa.com/",
                                        "user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36"})
        f = urllib.request.urlopen(req).read().decode("utf-8")

        soup = BeautifulSoup(f, "html.parser")

        review_num = len(soup.find_all("div", {"class":"review-contents__text"}))

        for i in range(review_num):
            review = soup.find_all("div", {"class":"review-contents__text"})[i].text
            star = soup.find_all("span", {"class":"review-list__rating__active"})[i]["style"][-4:]
            if star == '100%':
                star = '5'
            elif star == " 80%":
                star = '4'
            elif star == " 60%":
                star = '3'
            elif star == " 40%":
                star = '2'
            elif star == " 20%":
                star = '1'

            reviews.append([goodsNo, star, review])
    except:
        break
```

```python
# 결과를 데이터프레임으로 저장

import pandas as pd
from datetime import datetime

date = datetime.today().strftime('%Y-%m-%d')

results_df = pd.DataFrame(reviews)
results_df.columns = ['goodsNo', 'stars', 'review']
results_df.to_excel(date + ' Musinsa crawling.xlsx')
```
